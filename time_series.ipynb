*Below Graph Describes The Sales Based On The Transaction Of 3 Months (M1,M2,M3)*

```python
plt.figure(figsize = (12,6));
plt.title("The Sales Based On The Transaction Of 3 Months");
sns.barplot(x = 'MONTH',y='VALUE',data=data ,color='maroon');
```

*By Visualizing The above Graph We Can Conclude That Month 3 Has More Transactions Than The Remaining.*

*Above Graph Shows The Mapping Between Quantity and value For The Product.*

*Plotting The Each Store With Quantity.*

```python
plt.figure(figsize = (12,6))
sns.barplot(x = 'STORECODE',y='QTY',color="green",data=data);
```

*Plotting The Each Store With Value.*

```python
plt.figure(figsize = (12,6))
sns.barplot(x = 'STORECODE',y='VALUE',data=data)
```
